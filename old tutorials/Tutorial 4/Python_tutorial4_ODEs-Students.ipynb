2. Partial differential equations (PDE's): hyperbolic, elliptic, and parabolic

$$ a \frac{\partial \phi}{\partial x} + b\frac{\partial \phi}{\partial y} + q(x,y) \phi= F(x,y) \quad --> {\rm First\ order\ PDE's} $$

$$ a \frac{\partial^2 \phi}{\partial^2 x} + 2b\frac{\partial^2 \phi}{\partial x \partial y} + c\frac{\partial^2 \phi}{\partial^2 y} =0 \quad --> {\rm Second\ order\ PDE's} $$

Some specific equations that are very important in physics:


$$ {\rm Laplace\ equation}\qquad \quad \nabla^2 \Psi =0 \qquad {\rm elliptic} $$
$$ {\rm Poisson\ equation}\qquad \nabla^2 \Psi  = \frac{\rho}{\epsilon_{0}} \qquad {\rm elliptic} $$
$$ {\rm Wave\ equation}\quad \nabla^2 \Psi  = \frac{1}{c^2}\frac{\partial^2 \Psi}{\partial^2 t} \quad {\rm hyperbolic} $$

$$ {\rm Difussion\ equation}\quad \nabla^2 \Psi  = a\frac{\partial \Psi}{\partial t} \quad {\rm parabolic} $$

4. Convergence and self-convergence test $Q$

$$ k^{Q} \equiv \frac{E_{1}}{E_{2}} = \frac{y_{1}-y_{2}}{y_{2}-y_{3}}\quad {\rm where}\quad k = \frac{dx_{1}}{dx_{2}}\quad {\rm --> self-convergence }$$

$$ k^{Q} \equiv \frac{E_{1}}{E_{2}} = \frac{y_{1}-y_{0}}{y_{2}-y_{0}}\qquad {\rm where}\  y_{0}\ {\rm exact\ is\ the\ solution}\quad {\rm --> convergence }$$

$y_{1},\ y_{2}$, and $y_{3}$ are our solution with resolution $dx_{1},\ dx_{2}$ and $dx_{3}$ respectively. We usually use $dx_{2}=dx_{1}/2$ and $dx_{3}=dx_{2}/2$, so, we have $k=2$. See for example this <a href='http://www.scielo.org.mx/pdf/rmfe/v56n1/v56n1a7.pdf'>reference</a>, and <a href='https://github.com/EXPLORE-for-students/python-tutorials/blob/main/Tutorial%203/jupyter_notebook_students.ipynb'>Julia's</a> tutorial.

- Change resolution in <a href='https://numpy.org/doc/stable/reference/generated/numpy.linspace.html'>linspace</a> --> random distributed
- Change resolution in <a href='https://numpy.org/doc/stable/reference/generated/numpy.arange.html'>arange</a>     --> spaced by $dx$, this is better if we want to compute a convergence test


# Example 1: First Order ODE's
(Initial condition problem)


$$\frac{dy}{dx} = y_{0}-k*y, \quad y_{0}=y(x=0), \quad k=cte $$

```python
import numpy as np
from scipy.integrate import odeint

# function that returns the righ hand side (rhs)  of the ODE dy/dt 
def rhs(y,x):
    k = 0.3
    dydt = -k * y
    return dydt
#------------------

# initial condition
y0 = 3

# time array
#t = np.linspace(0,25)
dx=0.5
x = np.arange(0, 25,dx)

# solve ODE
y = odeint(rhs,y0,x)

# plot results
fig = plt.figure(figsize=(8,6))
plt.plot(x,y, 'r')
plt.xlabel(r'$\rm x$')
plt.ylabel(r'$\rm y$')
plt.show()
```

# Self convergence test

```python
# -------- Resolution 1
Nx = 50
#x1 = np.linspace(0,20,num=Nx)
dx=0.5
x1 = np.arange(0, 25,dx)


y1 = odeint(rhs,y0,x1)  # solve ODE
print(len(x1),len(y1))


# -------- Resolution 2
Nx = 100
#x2 = np.linspace(0,20,num=Nx)
dx=0.5/2.
x2 = np.arange(0, 25,dx)
y2 = odeint(rhs,y0,x2)  # solve ODE
#print(len(x2),len(y2))

# -------- Resolution 3
Nx = 200
#x3 = np.linspace(0,20,num=Nx)
dx=0.5/4.0
x3 = np.arange(0, 25,dx)

y3 = odeint(rhs,y0,x3)  # solve ODE
#print(len(t3),len(y3))


# --------  Error 
Error1=[]
Error2=[]

for i in range(len(x1)):
    for j in range(len(x2)):
        if x2[j]==x1[i]:
            #print(t1[i])
            E = y1[i] - y2[j]
            Error1.append(E)
print(len(Error1))

for i in range(len(x1)):
    for j in range(len(x2)):
        for k in range(len(x3)):
            if x2[j]==x1[i] and x2[j]==x3[k]:
                #print(t1[i])
                E = y2[i] - y3[j]
                Error2.append(E)

#print(len(Error2))


#for i in range(len(Error2)):
#    Error2[i]=4.0 * Error2[i]

# Order of convergence
# plot results
plt.plot(x1,Error1, color='blue', label=r'$E_{1}$')
plt.plot(x1,Error2, color='red', label=r'$E_{2}$')
#plt.plot(t1,Error2, color='red', label=r'$4 E_{2}$')
plt.legend(loc='upper right', ncol=2, fontsize=12, frameon=False)

#plt.yscale('log')

plt.xlabel(r'$\rm x$', fontsize=15)
plt.ylabel(r'$\rm Error$')
plt.show()
```

```python
Q=[]
for i in range(len(x1)):
    k = np.abs(Error1[i]/Error2[i])
    q = np.log(k)/np.log10(2.)
    Q.append(q)
    
# plot results
plt.plot(x1,Q, color='blue', label=r'$E_{1}$')
plt.legend(loc='upper right', ncol=2, fontsize=12, frameon=False)

plt.xlabel(r'$\rm t$')
plt.ylabel(r'$\rm Q$')
plt.show()
```

```python
# Using the L-norm
L21 =np.linalg.norm(Error1)**2
L22 =np.linalg.norm(Error2)**2
q = np.log10(L21/L22)/np.log10(2.)
print("Self-convergence test is %1.2f" % q)
```

# Example 2: Second Order PDE's
## Driven damped pendulum

$$\frac{d^2 \theta}{d^2 t} + \frac{1}{Q} \frac{d \theta}{d t} -\sin (\theta) - d \cos (\Omega t) =0,$$

where:
$Q$ is the quality factor,  $d$ is the forcing amplitude, $\Omega$ is the driving frequency of the forcing, $Q$ and $d$ absorved $g$, the mass $m$ and length fo the pendulum $L$.


We can transform our second order ODE into two coupled first order ODEs by introducing $\omega \equiv d\theta/dt$

$$\frac{d \theta}{d t} =\omega,\quad
\frac{d \omega}{d t} =-\frac{1}{Q} \omega +\sin (\theta) + d \cos (\Omega t) ,$$


Initial value and parameters:

$Q=2.0,\ d=1.5,\ \Omega=0.65$

$\theta(t=0)=0.0,\quad \omega(t=0)=0.0$



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def rhs(y, t, params):
    theta, omega = y      # unpack current values of y
    Q, d, Omega = params  # unpack parameters
    derivs = [omega,      # list of dy/dt=f functions
             -omega/Q + np.sin(theta) + d*np.cos(Omega*t)]
    return derivs

# Parameters
Q = 2.0          # quality factor (inverse damping)
d = 1.5          # forcing amplitude
Omega = 0.65     # drive frequency

# Initial values
theta0 = 0.0     # initial angular displacement
omega0 = 0.0     # initial angular velocity

# Parameters for ODE solver
params = [Q, d, Omega]

# Initial conditions for ODE solver
y0 = [theta0, omega0]

# Make time array for solution
tStop = 200.      # final time
tInc = 0.05       # time step size
t = np.arange(0., tStop, tInc)

# Call the ODE solver
psoln = odeint(rhs, y0, t, args=(params,))

# Plot results
fig = plt.figure(1, figsize=(15,10))

# Plot theta as a function of time
ax1 = fig.add_subplot(311)
ax1.plot(t, psoln[:,0], color='blue')
ax1.set_xlabel(r'$t$')
ax1.set_ylabel(r'$\theta$')

# Plot omega as a function of time
ax2 = fig.add_subplot(312)
ax2.plot(t, psoln[:,1], color='red')
ax2.set_xlabel(r'$t$')
ax2.set_ylabel(r'$\omega$')

# Plot omega vs theta
ax3 = fig.add_subplot(313)
twopi = 2.0*np.pi
ax3.plot(psoln[:,0]%twopi, psoln[:,1], '.', ms=1, color='green', label='Phase space diagram')
ax3.set_xlabel(r'$\theta$')
ax3.set_ylabel(r'$\omega$')
ax3.set_xlim(0., twopi)

plt.tight_layout()
plt.show()
```
